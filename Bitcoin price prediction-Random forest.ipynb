```python
import pandas as pd
data=pd.read_csv("data9.csv",sep=",")
data
```

```python
data['date'] = pd.to_datetime(data['date']) 
data.sort_values(by=['date'], inplace=True)
data
```

```python
data['Nasdaq composite index']=data['Nasdaq composite index'].fillna(method='ffill')
data['DJI']=data['DJI'].fillna(method='ffill')
data['Nasdaq composite index']=data['Nasdaq composite index'].fillna(method='bfill')
data['DJI']=data['DJI'].fillna(method='bfill')
data['Gold in USD']=data['Gold in USD'].fillna(method='ffill')
data['Gold in USD']=data['Gold in USD'].fillna(method='bfill')
data
```

```python
data.set_index(['date'],inplace=True)
data
```

```python
split_point = int(len(data)*0.8)
split_point
```

```python
label = data['BTC Price']
feature= data.drop(['BTC Price'], axis = 1)
```

```python
train_f = feature.iloc[:split_point,:].values
train_l = label.iloc[:split_point].values
test_f = feature.iloc[split_point:,:].values
test_l = label.iloc[split_point:].values
```

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
#labels = np.array(data['BTC Price'])
#features= data.drop(['BTC Price'], axis = 1)
f_list = list(feature.columns)
#features = np.array(features)
#train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.2, random_state = 42)
print('Training Features Shape:', train_f.shape)
print('Training Labels Shape:', train_l.shape)
print('Testing Features Shape:', test_f.shape)
print('Testing Labels Shape:', test_l.shape)
```

```python
rf = RandomForestRegressor(max_depth=110,n_estimators = 1000, max_features='sqrt', random_state = 42)
rf.fit(train_f, train_l);
pred = rf.predict(test_f)
err = abs(pred - test_l)
print('Mean Absolute Error:', round(np.mean(err), 2), 'degrees.')
mape = 100 * (err / test_l)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
# Calculate the absolute errors
err = np.sqrt(np.mean(np.square(pred.reshape(-1,1) - test_l.reshape(-1,1) )))
# Print out the root mean square error (rmse)
print('RMSE:', err)
```
